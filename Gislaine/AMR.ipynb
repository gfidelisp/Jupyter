```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*
import csv

rc('text',usetex = True)
rc('font', family='serif')
```

```python
dT = [
    [18.99853892,22.16083315,25.34208814,28.5491296],
    [19.30115049,22.37010885,25.45558083,28.56558879],
    [19.0195731,22.09667285,25.19695865,28.33169549],
    [18.48459534,21.74049616,25.01988047,28.33133607] 
]

Qc = [
    [20.75974568,17.39655154,13.6413343,9.35110917],
    [16.33264639,14.7286914,12.73743461,10.16936026],
    [22.85921451,21.07362598,18.74504969,15.60827407],
    [30.3446981,25.22087504,19.62786967,13.39312684]
]

T = arange(0,30)
```

```python
fig, ax = plt.subplots()

plot(dT[1],Qc[1],'^-',label = '$\phi$ = 0.4 - FE = 50 \%')
plot(dT[0],Qc[0],'s-',label = '$\phi$ = 0.6 - FE = 50 \%')
plot(dT[2],Qc[2],'o-',label = '$\phi$ = 0.4 - FE = 35 \%')
plot(dT[3],Qc[3],'x-',label = '$\phi$ = 0.6 - FE = 35 \%')

plot(T,0.5*T,'--',label = '$UA_{gab}$ = 0.5 W/K')
plot(T,1*T,'--',label = '$UA_{gab}$ = 1 W/K')
plot(T,1.6*T,'--',label = '$UA_{gab}$ = 1.6 W/K')

ylabel('Cooling Capacity [W]')
xlabel('Temperature Span [K]')

xlim(18,29)
ylim(8,32)
plt.legend(bbox_to_anchor=(0.4, 1, 1, 0.02))
grid(True)

show()

fig.savefig('Qc_AMR.png',format = 'png', dpi=500,bbox_inches='tight')
```
