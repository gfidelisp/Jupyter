```python
def eta2(X,a,b,c,d,e,f,g,h,i,j): 
    Pi1,Pi2,Pi3,Pi4,Pi5,Pi6,Pi7 = X
    return (a*Pi1 + b*(Pi2) + c*(Pi3)+d*Pi4+e*Pi5+f*Pi6+g*Pi7)
```

```python
def eta(X,a,b,c,d,e,f,g,h,i,j,k): 
    Pi1,Pi2,Pi3,Pi4,Pi5,Pi6,Pi7,Pi8,Pi9,Pi10 = X
    return (a*(Pi1**b)*(Pi2**c)*(Pi3**d)*(Pi4**e)*(Pi5**f)*(Pi6**g)*(Pi7**h)*((Pi8*Pi9)**i)*(Pi9**j)*(Pi10**k))
```

```python
def Vi(x,y,i):
    Sum = 0
    for n in range(i-1,100):
        Sum = Sum + (binom(n,i-1)*(y/x)**(n/2)*iv(n,2*(x*y)**0.5)) 
    V = exp(-(x+y))*Sum
    return(V)
```

```python
def Betai(NTU,Cr,i):
    Beta = Vi(2*NTU,2*NTU/Cr,i)/(2*NTU)**(i-1)
    return Beta
```
