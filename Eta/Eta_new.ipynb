```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*
import csv
import xlrd

rc('text',usetex = True)
rc('font', family='serif')

from scipy.optimize import curve_fit
```

```python
def eta(X,a,b,c,d,e,f,g,h,i): 
    Pi1,Pi2,Pi3,Pi4 = X
    return (a*(Pi1**b)*(Pi2**c)*(Pi3**d)*(Pi4**e)) #+f*np.exp(Pi1)+g
```

```python
workbook = xlrd.open_workbook('Data.xlsx')
```

```python
worksheet = workbook.sheet_by_name('Data')
```

```python
N = 473

phi = zeros(N)
L = zeros(N)
D = zeros(N)
ha = zeros(N)
NTU = zeros(N)
Qc = zeros(N)
Qcf = zeros(N)
f = zeros(N)
m = zeros(N)
mf = zeros(N)

for i in range(N):
    phi[i] = worksheet.cell(0, i).value
    L[i] = worksheet.cell(1, i).value
    D[i] = worksheet.cell(2, i).value
    ha[i] = worksheet.cell(3, i).value
    NTU[i] = worksheet.cell(4, i).value
    Qc[i] = worksheet.cell(5, i).value
    Qcf[i] = worksheet.cell(6, i).value
    f[i] = worksheet.cell(7, i).value
    m[i] = worksheet.cell(8, i).value
    mf[i] = worksheet.cell(8, i).value
```

```python
M= curve_fit(eta,(phi,L,D,ha), Qc,bounds=([-1,-1,-0.1,-1,-0.1,-1000,-10,-10,-10], [1000,10,1,0.1,0.1,1,100,100,100]))
#M = curve_fit(eta,phi, Qc,bounds=([-100,-100], [100,100]))
M[0]
```

```python
plt.plot(phi,Qc,'o')
plt.plot(phi,eta((phi,L,D,ha),M[0][0],M[0][1],M[0][2],M[0][3],M[0][4],M[0][5],M[0][6],M[0][7],M[0][8]),'o')
#plt.plot(phi,eta((phi),M[0][0],M[0][1]),'o')
plt.grid()
```
