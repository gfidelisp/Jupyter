```python
import matplotlib.pyplot as plt
import numpy as np
#from __future__ import division
from numpy import *
import xlrd
plt.rc('text',usetex = True)
plt.rc('font', family='serif')
```

```python
UA = 1.75
cfm = np.array([60,65,72,79,48,51])*1.1 #60
m = cfm*0.00047194745*1.225
cp = 1000
E = np.array([0.463,0.226,0.423,0.20,0.513,0.2615])*1.1 #0.462
T_w = np.array([-5,-3,-1,0,1,3,5]) #np.arange(-5,6,2)
W = [4.5,5.5,4.3,5.1,3.8,4.8]
```

```python
Data = xlrd.open_workbook('Results_experiments.xlsx')
HEx = Data.sheet_by_name('Data1')
```

```python
T_w_in_1  = array(HEx.col_values(0)) 
T_w_in_2  = array(HEx.col_values(1)) 
T_w_in_3  = array(HEx.col_values(2)) 
T_w_in_4  = array(HEx.col_values(3)) 
T_w_in_5  = array(HEx.col_values(4)) 
T_w_in_6  = array(HEx.col_values(5)) 

T_cab_1  = array(HEx.col_values(6)) 
T_cab_2  = array(HEx.col_values(7)) 
T_cab_3  = array(HEx.col_values(8)) 
T_cab_4  = array(HEx.col_values(9)) 
T_cab_5  = array(HEx.col_values(10)) 
T_cab_6  = array(HEx.col_values(11)) 

T_amb_1  = array(HEx.col_values(12)) 
T_amb_2  = array(HEx.col_values(13)) 
T_amb_3  = array(HEx.col_values(14)) 
T_amb_4  = array(HEx.col_values(15)) 
T_amb_5  = array(HEx.col_values(16)) 
T_amb_6  = array(HEx.col_values(17)) 
```

```python
T_cab_1
```

```python
T_erro = array([-3,10])
```

```python
T_cab_mod_1 = np.array((UA*np.array(25)+E[0]*cp*m[0]*np.array(T_w)+W[0])/(UA+E[0]*cp*m[0]))
T_cab_mod_2 = np.array((UA*np.array(25)+E[1]*cp*m[1]*np.array(T_w)+W[1])/(UA+E[1]*cp*m[1]))
T_cab_mod_3 = np.array((UA*np.array(25)+E[2]*cp*m[2]*np.array(T_w)+W[2])/(UA+E[2]*cp*m[2]))
T_cab_mod_4 = np.array((UA*np.array(25)+E[3]*cp*m[3]*np.array(T_w)+W[3])/(UA+E[3]*cp*m[3]))
T_cab_mod_5 = np.array((UA*np.array(25)+E[4]*cp*m[4]*np.array(T_w)+W[4])/(UA+E[4]*cp*m[4]))
T_cab_mod_6 = np.array((UA*np.array(25)+E[5]*cp*m[5]*np.array(T_w)+W[5])/(UA+E[5]*cp*m[5]))
```

```python
fig, ax = plt.subplots()

plt.plot(T_w_in_5,T_cab_5,'or',label = 'HEx 1')
plt.plot(T_w_in_3,T_cab_3,'ob',label = 'HEx 2')
plt.plot(T_w_in_1,T_cab_1,'ok',label = 'HEx 3')
plt.plot(T_w_in_6,T_cab_6,'oy',label = 'HEx 4')
plt.plot(T_w_in_4,T_cab_4,'og',label = 'HEx 5')
plt.plot(T_w_in_2,T_cab_2,'om',label = 'HEx 6')

plt.plot(T_w,T_cab_mod_5,'--r')
plt.plot(T_w,T_cab_mod_3,'--b')
plt.plot(T_w,T_cab_mod_1,'--k')
plt.plot(T_w,T_cab_mod_6,'--y')
plt.plot(T_w,T_cab_mod_4,'--g')
plt.plot(T_w,T_cab_mod_2,'--m')
plt.grid()
plt.ylabel('$T_\mathrm{Cab}$ [$^{\circ}$C]')
plt.xlabel('$T_\mathrm{w,in}$ [$^{\circ}$C]')
plt.legend()

fig.savefig('T_cab_new.png',format = 'png', dpi=1200,bbox_inches='tight')
```
