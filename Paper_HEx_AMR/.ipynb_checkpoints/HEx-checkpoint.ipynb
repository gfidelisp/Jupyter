```python
x = np.arange(0.1,501.01,1)
```

```python
Cf = 150/3600*4148

Ca = np.zeros(len(x))
NTU = np.zeros([len(x),6])
Cr = np.zeros(len(x))
K1 = np.zeros([len(x),6])
E = np.zeros([6,len(x)])
E1 = np.zeros([6,len(x)])
ECr = np.zeros([6,len(x)])
QECr = np.zeros([6,len(x)])
QW = np.zeros([6,len(x)])

for i in range(len(x)):
    for j in range(6):
        Ca[i] = x[i]*0.00047194745*1.225*1005
        NTU[i][j] = UA_array[j][0][i]/Ca[i]
        Cr[i] = Ca[i]/Cf
        K1[i][j] = 1-np.exp(-NTU[i][j]/2)
        if j<3:
            E[j][i] = 1-np.exp(-(1-np.exp(-Cr[i]*NTU[i][j]))/Cr[i])
        else:    
            E[j][i] = (1-np.exp(-2*K1[i][j]*Cr[i])*(1+Cr[i]*K1[i][j]**2))/Cr[i]
        ECr[j][i] = E[j][i]*Cr[i]
        QECr[j][i] = M[0][0]*np.log(ECr[j][i]) + M[0][1]
        QW[j][i] =  QECr[j][i]-W_array[j][0][i]
```

```python
cfm,UA,W,dP

E_2 = np.zeros(len(UA))
Cr_2 = np.zeros(len(UA))
Ca_2 = np.zeros(len(UA))
NTU_2 = np.zeros(len(UA))
K1_2 = np.zeros(len(UA))
ECr_2 = np.zeros(len(UA))
QECr_2 = np.zeros(len(UA))
QW_2 = np.zeros(len(UA))

for i in range(len(UA)):
    Ca_2[i] = cfm[i]*0.00047194745*1.225*1005
    NTU_2[i] = UA[i]/Ca_2[i]
    Cr_2[i] = Ca_2[i]/Cf
    K1_2[i] = 1-np.exp(-NTU_2[i]/2)
    if j<3:
        E_2[i] = (1-np.exp(-(1-np.exp(-Cr_2[i]*NTU_2[i]))/Cr[i]))
    else:
        E_2[i] = (1-np.exp(-2*K1_2[i]*Cr_2[i])*(1+Cr_2[i]*K1_2[i]**2))/Cr_2[i]
    ECr_2[i] =  E_2[i]*Cr_2[i]
    QECr_2[i] = M[0][0]*np.log(ECr_2[i]) + M[0][1]
    QW_2[i] =  QECr_2[i]-W[i]
```
