```python
from numpy import *
from scipy import *
import numpy as np
from matplotlib import *

from scipy import integrate
from scipy import interpolate
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import xlrd

rc('text',usetex = True)
rc('font', family='serif',size = 13)
```

```python
Data = xlrd.open_workbook('Data.xlsx')
Inputs = Data.sheet_by_name('Plan1')

phi   = array(Inputs.col_values(0)) 
B1  = array(Inputs.col_values(1)) 
B2   = array(Inputs.col_values(2))
B3 = array(Inputs.col_values(4))
```

```python
plt.plot(phi,B1,':',label = 'Com rampa - B$_\mathrm{alvo}$ = 1.1 T  ')
plt.plot(phi,B2,'-.',label = 'Com rampa - B$_\mathrm{alvo}$ = 1.8 T')
plt.plot(phi,B3,'--',label = 'Sem rampa - B$_\mathrm{alvo}$ = 1.8 T')
plt.grid()
plt.xlabel(r'$\theta$ [$^{\circ}$]')
plt.legend()
plt.xlim([0,90])
plt.ylim([0,1.6])
plt.ylabel('B [T]')

plt.savefig('Campo.png',format = 'png', dpi=500,bbox_inches='tight')
```
