# Volumetric efficiency of magnet systems and application to higher capacities

## Magnet volumetric efficiency

Consider a magnet of volume $\mathcal{V}_{\mathrm{M}}$, composed of a material of uniform permeability $\mu_{\mathrm{r,M}}$ and uniform remanence $B_{\mathrm{rem}}$. This magnet must generate a field $\vec{H}$ over a field region of volume $\mathcal{V}_{\mathrm{F}}$.

From Jensen and Abele (1996), the energy of the total system $U$ is expressed as:

$$
U = \frac{\mu_0}{2} \int_{\mathcal{V}_{\mathrm{F}}} H^2 \mathrm{d}\Omega + \frac{\mu_0 \mu_{\mathrm{r,M}}}{2} \int_{\mathcal{V}_{\mathrm{M}}} H^2 \mathrm{d}\Omega
$$

where the first term in the energy of the field region $U_{\mathrm{F}}$ and the second, the energy of the magnet region $U_{\mathrm{M}}$.

According to Jensen and Abele (1996), the magnet energy has an upper limit $U_0$, in the case the remanence is optimally distributed, calculated as:

$$
U_0 = \frac{B_{\mathrm{rem}}^2 \mathcal{V}_{\mathrm{M}}}{2 \mu_0 \mu_{\mathrm{r,M}}}
$$

It can be shown (Jensen and Abele, 1996) that there is an upper limit to the magnetic energy that the magnet can generate external to itself, given by:

$$
U^* = \frac{U_0}{4} = \frac{B_{\mathrm{rem}}^2 \mathcal{V}_{\mathrm{M}}}{8 \mu_0 \mu_{\mathrm{r,M}}}
$$

This means that the energy that is distributed to the field region is constrained by this upper limit:

$$
U_{\mathrm{F}} \le U^*
$$

Suppose that field over this region is uniform and equal to $H_{\mathrm{h}}$; then

$$
\frac{\mu_0 H_{\mathrm{h}}^2}{2} \mathcal{V}_{\mathrm{F}} \le \frac{B_{\mathrm{rem}}^2 \mathcal{V}_{\mathrm{M}}}{8 \mu_0 \mu_{\mathrm{r,M}}}
$$

Multiplying both sides by $\frac{8\mu_0\mu_{\mathrm{r,M}}}{B_{\mathrm{rem}^2}}$ and letting $B_{\mathrm{h}} = \mu_0 H_{\mathrm{h}}$, we have:

$$
\mathcal{V}_{\mathrm{M}} \ge 4 \mu_{\mathrm{r,M}} \left(\frac{B_{\mathrm{h}}}{B_{\mathrm{rem}}} \right)^2 \mathcal{V}_{\mathrm{F}}
$$

Now, consider the total gap volume $\mathcal{V}_{\mathrm{gap}}$; the field region is half of that volume, because the other half is demagnetized (Bj√∏rk et al., 2011). Therefore:

$$
\mathcal{V}_{\mathrm{M}} \ge 2 \mu_{\mathrm{r,M}} \left(\frac{B_{\mathrm{h}}}{B_{\mathrm{rem}}} \right)^2 \mathcal{V}_{\mathrm{gap}}
$$

The critical magnet volume, i.e. the minimum magnet volume that can generate the high field over half the gap volume is:

$$
\mathcal{V}_{\mathrm{M}}^* = 2 \mu_{\mathrm{r,M}} \left(\frac{B_{\mathrm{h}}}{B_{\mathrm{rem}}} \right)^2 \mathcal{V}_{\mathrm{gap}}
$$

This yields a natural definition of a *magnet volumetric efficiency* $\nu$:

$$
\nu = \frac{\mathcal{V}_{\mathrm{M}}^*}{\mathcal{V}_{\mathrm{M}}}
$$

where $0 \le \nu \le 1$. What causes $\nu < 1$?

* The magnet material is not ideally magnetized, i.e. the remanence field distribution is not optimal
* Some of the magnetic flux generated by the magnet is leaking to regions outside the field region
* Some parts of the magnet region are demagnetized

In addition, notice that $\mathcal{V}_{\mathrm{M}}^* \propto \mu_{\mathrm{r,M}}$; by having a material with relative permeability higher than unity, we are being penalized and must have a larger magnet.


## Functions

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

import nemplot
from nemplot import nemplot_parameters
import teslamax
```

```python
def calculate_critical_volume(mu_r,B_h,B_rem,V_gap):
    """
    Return the critical magnet volume for a magnet material with
    permeability 'mu_R' and remanence 'B_rem' generate a field of 'B_h'
    over half a gap volume 'V_gap'
    """
    
    return 2 * mu_r * (B_h/B_rem)**2 * V_gap
```

```python
def calculate_gap_volume(magnet_params):
    """
    Return the air gap volume for a magnet geometry described by 'magnet_params'
    """
    
    R_o = magnet_params["R_o"]
    h_gap = magnet_params["h_gap"]
    
    R_g = R_o + h_gap
    
    A = np.pi * (R_g**2 - R_o**2)
    
    L = magnet_params["L"]
    V = A*L
    
    return V
```

## Application to the TeslaMax model, no internal magnet, for higher capacities

Consider the TeslaMax model:

```python
from IPython.display import Image, display, HTML
Image("figures/teslamax.png",width=500)
```

For simplicity, consider the case of no internal magnet, as our recent simulations have shown it's worth it.

Let's calculate the magnet efficiency for the best compact case:

```python
# Fixed parameters

B_REM = 1.4
MU_R_MAGNET = 1.05
DENSITY_MAGNET = 7500

FIXED_PARAMETERS_AMR = {
    "D_p[m]": 0.35e-3,
    "L[m]": 85e-3,
    "W[m]": 30e-3,
    "H[m]": 17e-3,
    "Casing material": "Stainless steel",
    "t_casing[m]": 0.5e-3,
    "t_air[m]": 1e-3}

FIXED_PARAMETERS_COMPACT_MAGNET = {"R_i": 0.015,
                "h_fc": 0.005,
                "L": 1.1*FIXED_PARAMETERS_AMR["L[m]"],
                'h_gap': FIXED_PARAMETERS_AMR["H[m]"] + 2*(FIXED_PARAMETERS_AMR["t_casing[m]"] + FIXED_PARAMETERS_AMR["t_air[m]"]),
                "R_e": 0.3,
                "R_o": 40e-3,
                "R_s": 105e-3,
                "n_II": 0,
                "n_IV": 4,
                "phi_C_II": 0,
                "phi_S_II": 0,
                "phi_S_IV": 45,
                "mu_r_II": MU_R_MAGNET,
                "mu_r_IV": MU_R_MAGNET,
              "linear_iron": 1,
              "mu_r_iron": 5e3,
             }
```

```python
B_MAX_COMPACT = 1.1

V_critical_compact = calculate_critical_volume(B_h=B_MAX_COMPACT,
                                               B_rem=B_REM,
                                               mu_r=MU_R_MAGNET,
                                               V_gap=calculate_gap_volume(FIXED_PARAMETERS_COMPACT_MAGNET)
                                                )
print(V_critical_compact)
```

Let's calculate the actual magnet volume and the efficiency:

```python
tm_geometry_compact = teslamax.TeslaMaxGeometry(FIXED_PARAMETERS_COMPACT_MAGNET)
```

```python
V_compact = tm_geometry_compact.calculate_magnet_volume(L=FIXED_PARAMETERS_COMPACT_MAGNET["L"])
print(V_compact)
```

```python
magnet_efficiency_compact = V_critical_compact / V_compact
print(magnet_efficiency_compact*100)
```

Now, for the larger magnet, the AMR and magnet parameters are given by:

```python
B_MAX_LARGE = 1.4

FIXED_PARAMETERS_AMR = {
    "D_p[m]": 0.35e-3,
    "L[m]": 120e-3,
    "W[m]": 30e-3,
    "H[m]": 30e-3,
    "Casing material": "Stainless steel",
    "t_casing[m]": 0.5e-3,
    "t_air[m]": 1e-3}

FIXED_PARAMETERS_LARGE_MAGNET = {"R_i": 0.015,
                "h_fc": 0.005,
                "L": 1.1*FIXED_PARAMETERS_AMR["L[m]"],
                'h_gap': FIXED_PARAMETERS_AMR["H[m]"] + 2*(FIXED_PARAMETERS_AMR["t_casing[m]"] + FIXED_PARAMETERS_AMR["t_air[m]"]),
                "R_e": 0.3,
                "n_II": 0,
                "n_IV": 4,
                "phi_C_II": 0,
                "phi_S_II": 0,
                "phi_S_IV": 45,
                "mu_r_II": MU_R_MAGNET,
                "mu_r_IV": MU_R_MAGNET,
              "linear_iron": 1,
              "mu_r_iron": 5e3,
             }

n_II = FIXED_PARAMETERS_LARGE_MAGNET["n_II"]
n_IV = FIXED_PARAMETERS_LARGE_MAGNET["n_IV"]

n = n_II + n_IV

# expand parameters to include remanence magnitudes for each segment
FIXED_PARAMETERS_LARGE_MAGNET = teslamax.expand_parameters_from_remanence_array(B_REM*np.ones(n), 
                                                                      FIXED_PARAMETERS_LARGE_MAGNET, 
                                                                      "B_rem")
```

We are interested in finding the minimum magnet volume that can yield a cooling capacity $\dot{Q}_{\mathrm{c}}$

Using AMR simulations with the above geometry, we find that we can select a frequency and an utilization factor such that the cooling capacity of each regenerator is $\dot{Q}_{\mathrm{c,reg}}$. The required number of regenerators is then:

$$
N_{\mathrm{r}} = \frac{\dot{Q}_{\mathrm{c}}}{\dot{Q}_{\mathrm{c,reg}}}
$$

These regenerators will be distributed in the gap between cylinders. In trying to be as compact as possible, the minimum value of $R_{\mathrm{o}}$ that will yield a gap where the regenerator can fill the entire perimeter is given by:

$$
R_\mathrm{o} = \frac{N_{\mathrm{r}} W}{2\pi} - \frac{h_{\mathrm{gap}}}{2}
$$

and the gap volume will be:

$$
\mathcal{V}_{\mathrm{gap}} = \pi \left(R_\mathrm{g}^2 - R_\mathrm{o}^2\right) L_\mathrm{M}
$$

With this, we can calculate the minimum magnet volume for the desired capacity, assuming a given magnet volumetric efficiency

```python
NU = magnet_efficiency_compact
BTU_PER_HOUR_TO_KW = 293e-6

Q_c_values_BTUh = np.array([6000,9000,12000,15000,18000])
Q_c_reg_W = 100
B_rem_values = np.array([1.08,1.38])

h_gap = FIXED_PARAMETERS_LARGE_MAGNET["h_gap"]/2
W = FIXED_PARAMETERS_AMR["W[m]"]
L_M = FIXED_PARAMETERS_LARGE_MAGNET["L"]

magnet_volume_table = DataFrame()

for B_rem in B_rem_values:
    
    for Q_c_BTUh in Q_c_values_BTUh:
        
        Q_c_kW = Q_c_BTUh * BTU_PER_HOUR_TO_KW

        N_r = np.ceil(1e3*Q_c_kW / Q_c_reg_W).astype(int)
        R_o = (N_r * W)/(2*np.pi) - h_gap/2
        
        R_g = R_o + h_gap
        
        V_gap  = np.pi * (R_g**2 - R_o**2) * L_M
        
        V_magnet_critical = calculate_critical_volume(B_h=B_MAX_LARGE,
                                                      B_rem=B_rem,
                                                      mu_r=MU_R_MAGNET,
                                                      V_gap=V_gap)
        
        V_PMM = 1/NU * V_magnet_critical
        m_PMM = V_PMM * DENSITY_MAGNET
        
        s = Series({"Q_c[BTU/h]": Q_c_BTUh,
                   "B_rem[T]": B_rem,
                   "N_r": int(N_r),
                   "R_o[mm]": R_o*1e3,
                   "V_PMM[L]": V_PMM*1e3,
                   "m_PMM[kg]": m_PMM})
        
        magnet_volume_table = magnet_volume_table.append(s,ignore_index=True)
        
        
        
        
```

```python
def filter_table_from_column(table,column,value):
    """
    Return a view into the 'table' DataFrame, selecting only the rows where
    'column' equals 'value'
    """
    
    return table[table[column] == value]
```

```python
def get_large_magnet_vectors(B_rem):
    """
    Return (Q_c_BTUh, V_PMM_L,m_PMM_kg) as arrays to be plotted, 
    based on the volumetric efficiency calculations for large magnets for remanence 'B_rem'
    """
    
    map_table = filter_table_from_column(magnet_volume_table,'B_rem[T]',B_rem)

    Q_c = map_table["Q_c[BTU/h]"].values
    V_PMM = map_table["V_PMM[L]"].values
    m_PMM = map_table["m_PMM[kg]"].values
    
    return (Q_c,V_PMM,m_PMM)
```

```python
nemplot.set_figsize_cm(10)
nemplot.set_latex_font('Palatino')
nemplot.set_fontsize(11)
```

```python
V_PMM_MIN = 0
V_PMM_MAX = 20

def plot_large_magnet_volume():
    
        
    fig, ax,ax_m = nemplot.create_two_axes_plot(xlabel=r'\text{Capacidade de refrigera{\c c}{\~ a}o} $[\si{BTU\per\hour}]$',
                                                ylabel_left=r'\text{Volume de Nd-Fe-B} $[\si{\liter}]$',
                                                ylabel_right=r'\text{Massa de Nd-Fe-B} $[\si{\kg}]$'
                                               )
    
    
    Q_c_values = magnet_volume_table["Q_c[BTU/h]"].unique()
    for B_rem in B_rem_values:
        
        Q_c, V_PMM,m_PMM = get_large_magnet_vectors(B_rem)
        
        ax.plot(Q_c,V_PMM,'s--',label="Reman{\^ e}ncia %.2f T" %(B_rem))
        
        
    ax.grid(True)
    ax.legend(loc='upper left',fontsize=0.8*nemplot_parameters["FONTSIZE"])
    
    ax.set_xlim(np.min(Q_c_values),np.max(Q_c_values))
    ax.set_xticks(Q_c_values)
    
    ax.set_ylim(V_PMM_MIN,V_PMM_MAX)
    V_ticks = np.linspace(V_PMM_MIN,V_PMM_MAX,11)
    
    ax.set_yticks(V_ticks)
    
    nemplot.refine_yticks(ax,5)
    
    m_ticklabels = ["%.1f" %(DENSITY_MAGNET*1e-3*V,) for V in V_ticks]
    ax_m.set_yticks(V_ticks)
    ax_m.set_yticklabels(m_ticklabels)
    ax.set_yticks(V_ticks)
    nemplot.refine_yticks(ax_m,5)

    
    nemplot.save_figure(fig,"magnet_volume_large")
```

```python
%matplotlib inline
plot_large_magnet_volume()
```

Let's calculate the overall dimensions a large magnet would have for a sample cooling capacity.

Using only the external magnet, the volume is given by:

$$
\mathcal{V}_{\mathrm{PMM}} = 2 \phi_{\mathrm{S,IV}} \left(R_\mathrm{s}^2 - R_\mathrm{g}\right)^2 L
$$

and hence, for a given volume, the external radius is given by:

$$
R_\mathrm{s} = \left(R_\mathrm{g}^2 + \frac{\mathcal{V}_{\mathrm{PMM}}}{ 2 \phi_{\mathrm{S,IV}}L}\right)^{\frac{1}{2}}
$$

```python
Q_c = 9000
B_REM_LARGE = 1.38

table = filter_table_from_column(magnet_volume_table,"B_rem[T]",B_REM_LARGE)

table = filter_table_from_column(table,"Q_c[BTU/h]",Q_c)
R_o = 1e-3*table["R_o[mm]"].values[0]
V_PMM = 1e-3*table["V_PMM[L]"].values[0]

h_gap = FIXED_PARAMETERS_LARGE_MAGNET["h_gap"]

R_g = R_o + h_gap

phi_S_IV = np.deg2rad(FIXED_PARAMETERS_LARGE_MAGNET["phi_S_IV"])
L = FIXED_PARAMETERS_LARGE_MAGNET["L"]

R_s = np.sqrt(R_g**2 + (V_PMM)/(2*phi_S_IV*L))

print("For a capacity of %d BTU/h:" %(Q_c,))
print()
print("R_o = %.1f mm" %(R_o*1e3,))
print("R_g = %.1f mm" %(R_g*1e3,))
print("h_gap = %.1f mm" %(h_gap*1e3,))
print("R_s = %.1f mm" %(R_s*1e3,))
```
