```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*

import xlrd

matplotlib.rcParams.update({'font.size': 12})
rc('text',usetex = True)
rc('font', family='serif')
```

```python
Ang = [45,50,60,70,80]

B_max_with    = [1.4,1.45,1.5,1.45,1.45]
B_max_without = [1.3,1.3,1.35,1.3,1.3] 

B_min_with = [0,0,0,0.2,0.25]
B_min_without = [0,0,0,0.2,0.23]
```

```python
plt.plot(Ang,B_max_with,'o--', label = 'Configura\c{c}\~ao 1')
plt.plot(Ang,B_max_without,'o--', label = 'Configura\c{c}\~ao 2')
plt.xlabel('$\phi_\mathrm{S,IV}$ [$^{\circ}$]', fontsize = 12)
plt.ylabel('$\mu_0$ H$_\mathrm{max}$ [T]', fontsize = 12)
plt.grid()
plt.legend()
plt.ylim([1.275,1.525])
plt.savefig('Bmax.png',format = 'png', dpi=1000,bbox_inches='tight')
```

```python
plt.plot(Ang,B_min_with,'o--', label = 'Configura\c{c}\~ao 1')
plt.plot(Ang,B_min_without,'o--', label = 'Configura\c{c}\~ao 2')
plt.xlabel('$\phi_\mathrm{S,IV}$ [$^{\circ}$]', fontsize = 12)
plt.ylabel('$\mu_0$ H$_\mathrm{min}$ [T]', fontsize = 12)
plt.grid()
plt.legend()
plt.ylim([-0.0275,0.3])
plt.savefig('Bmin.png',format = 'png', dpi=1000,bbox_inches='tight')
```
