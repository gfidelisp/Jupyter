```python
import numpy as np
from matplotlib.pyplot import *
```

```python
H = np.arange(30,70,1)
W = np.arange(30,70,1)
L = np.zeros([len(H),len(W)])

m = 1.5
Epsilon  = 0.5
rho = 7000

for i in range(len(H)):
    for j in range(len(W)):
        L[i][j] = m/((1-Epsilon)*rho*10**-9*H[i]*W[j])
```

```python
CS = contour(W,H,L)
clabel(CS, inline=1, fontsize=10)
```
