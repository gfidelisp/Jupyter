```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*

import xlrd

matplotlib.rcParams.update({'font.size': 10})
rc('text',usetex = True)
rc('font', family='serif')

from scipy.optimize import curve_fit
from scipy.optimize import fsolve
```

```python
workbook = xlrd.open_workbook('Data.xlsx')
AMR_Data = workbook.sheet_by_name('AMR')
Magnet_Data = workbook.sheet_by_name('Magnet')
```

```python
N =287
B_t   = zeros(N)
h_gap = zeros(N)
H     = zeros(N)
W     = zeros(N)
Qc    = zeros(N)
Wp    = zeros(N)
Wm    = zeros(N)
R_4 = zeros(N)

for i in range(N):
    B_t[i]   = AMR_Data.cell(i+1, 0).value
    h_gap[i] = AMR_Data.cell(i+1, 1).value
    H[i] = AMR_Data.cell(i+1, 2).value
    W[i] = AMR_Data.cell(i+1, 3).value
    Qc[i] = AMR_Data.cell(i+1, 5).value
    Wp[i] = AMR_Data.cell(i+1, 6).value
    Wm[i] = AMR_Data.cell(i+1, 7).value    
```

```python
Qc_Target = 2650
```

```python
N_reg = ceil(Qc_Target/Qc)
N_reg = ceil(Qc_Target/Qc/4)*4
N_reg = (Qc_Target/Qc)
```

```python
plt.plot(N_reg,'o')
plt.ylim([0,16])
```

```python
R_2 = (N_reg*W/np.pi-h_gap)/2
R_3 = R_2 + h_gap
```

```python
N1 = 26

R2 = zeros(N1)
R3 = zeros(N1)
R4 = zeros(N1)
B = zeros(N1)

for i in range(N1):
    R2[i] = Magnet_Data.cell(i+1, 1).value
    R3[i] = Magnet_Data.cell(i+1, 2).value
    R4[i] = Magnet_Data.cell(i+1, 3).value
    B[i] = Magnet_Data.cell(i+1, 4).value
```

```python
def eta(X,a,b,c,d): 
    R1, R2, R3= X
    return (a*(R1**b)*(R2**c)*(R3**d))
```

```python
M = curve_fit(eta,(R2,R3,R4),B)
M[0]
```

```python
plt.plot(R4,B,'o')
plt.plot(R4,eta((R2,R3,R4),M[0][0],M[0][1],M[0][2],M[0][3]),'o')
```

```python
B,eta((R2,R3,R4),M[0][0],M[0][1],M[0][2],M[0][3])
```

```python
Erro1 = abs(B - eta((R2,R3,R4),M[0][0],M[0][1],M[0][2],M[0][3]))/B*100
sum(Erro1)
```

```python
plt.plot(R4,Erro1,'o')
```

```python
def f(R4,R2,R3,B,a,b,c,d):
    f = B - a*R2**b*R3**c*R4**d
    return f
```

```python
for i in range(N):
    R_4[i] = fsolve(lambda R4: f(R4,R_2[i],R_3[i],1.1*B_t[i],M[0][0],M[0][1],M[0][2],M[0][3]),R_3[i])
```

```python
m_gd = N_reg*H*W*0.15*(1-0.55)*7900*10**-6
m_Nd = np.pi*(((R_4+25)**2-R_3**2)+(R_2**2))*10**-6*(0.15+0.05)*7500 #25**2
m_w  = 3*(N_reg*H*W*0.15*(0.55)*1000*10**-6 + 2*1000*np.pi*((0.01588/2)**2)*(40*0.60+3))
plt.plot(m_w)
```
