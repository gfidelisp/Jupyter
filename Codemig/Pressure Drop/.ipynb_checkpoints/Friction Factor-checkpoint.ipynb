```python
Data = xlrd.open_workbook('Data_All.xlsx') #Importa a planilha "Valvulas" como "Data"
Inputs = Data.sheet_by_name('Planilha1') #Importa a aba "Planilha2" como Inputs
```

```python
vs = np.array(Inputs.col_values(6)) #Importa a coluna A
f = np.array(Inputs.col_values(7)) #Importa a coluna B
```
