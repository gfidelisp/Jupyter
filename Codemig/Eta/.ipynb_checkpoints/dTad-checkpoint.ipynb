```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*
import csv
import xlrd

rc('text',usetex = True)
rc('font', family='serif')

from scipy.optimize import curve_fit
```

```python
def eta(X,a,b,c,d,e,f,g,h,i): 
    Pi1,Pi2,Pi3,Pi4,Pi5,Pi6,Pi7,Pi8 = X
    return (a*(Pi1**b)*(Pi2**0)*(Pi3**0)*(Pi4**0)*(Pi5**0)*(Pi6**0)*(Pi7**0)*(Pi8**0))
```

```python
workbook = xlrd.open_workbook('Input.xlsm')
worksheet = workbook.sheet_by_name('Input')
```

```python
N = 134
phi = zeros(N)
L = zeros(N)
D = zeros(N)
CA_s= zeros(N)
CA_f= zeros(N)
ha = zeros(N)
dT = zeros(N)
Wp = zeros(N)
Qc = zeros(N)
Qcf = zeros(N)
dp = zeros(N)


for i in range(N):
    phi[i] = worksheet.cell(0, i).value
    L[i] = worksheet.cell(1, i).value
    D[i] = worksheet.cell(2, i).value
    ha[i] = worksheet.cell(5, i).value
    CA_s[i] = worksheet.cell(3, i).value
    CA_f[i] = worksheet.cell(4, i).value
    dT[i] = worksheet.cell(6, i).value
    Qc[i] = worksheet.cell(8, i).value
    Wp[i] = worksheet.cell(7, i).value
    Qcf[i] = worksheet.cell(9, i).value
    dp[i] = worksheet.cell(10, i).value
```

```python
plt.plot(phi,Qc,'o')
plt.grid()
plt.xlabel('$\phi$ [-]')
#plt.ylabel('$\dot{Q}_\mathrm{C}$ [W]')
plt.ylabel('$\eta$ [-]')
#plt.savefig('Eta.png',format = 'png', dpi=1000,bbox_inches='tight')
```

```python
M = curve_fit(eta,(phi,L,D/L,dp,dT,CA_s,CA_f,Wp), Qc,bounds=([-1,-5,-1,0,-4,-2,-1,-1,-2], [10,2,5,1,0,0,3,5,0]))
#M = curve_fit(eta,phi, Qc,bounds=([-100,-100], [100,100]))
M[0]
```

def eta(X,a,b,c,d,e,f,g,h,i):

    Pi1,Pi2,Pi3,Pi4,Pi5,Pi6,Pi7 = X
    
    return (a*(Pi1**b)*(Pi2**c)*(Pi3**d)*(Pi4**e)*(Pi5**f)*(Pi6**g)*Pi7**h)

```python
plt.plot(phi,Qc,'o',label = 'Numerical')
plt.plot(phi,eta((phi,L,D/L,dp,dT,CA_s,CA_f,Wp),M[0][0],M[0][1],M[0][2],M[0][3],M[0][4],M[0][5],M[0][6],M[0][7],M[0][8]),'o',label = 'Fitting')
plt.xlabel('$\phi$ [-]')
plt.ylabel('$\eta$ [-]')
plt.legend()
plt.grid()
plt.savefig('Fitting_Restricted_8_Reduced.png',format = 'png', dpi=1000,bbox_inches='tight')
```

```python
Erro_rel = (Qc-eta((phi,L,D/L,dp,dT,CA_s,CA_f,Wp),M[0][0],M[0][1],M[0][2],M[0][3],M[0][4],M[0][5],M[0][6],M[0][7],M[0][8]))/Qc*100
```

```python
plt.plot(phi,abs(Erro_rel),'o')
plt.xlabel('$\phi$ [-]')
plt.ylabel('Error [$\%$]')
plt.grid()

plt.savefig('Error_Restricted_8.png',format = 'png', dpi=1000,bbox_inches='tight')
```
