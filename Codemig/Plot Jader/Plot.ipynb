```python
from numpy import *
from matplotlib import *
from pylab import *
from scipy import *
from numpy import*
import csv
import xlrd
import matplotlib.pyplot as plt

rc('text',usetex = True)
rc('font', family='serif')

matplotlib.rcParams.update({'font.size': 12})

from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit
```

```python
workbook = xlrd.open_workbook('Input.xlsx')
worksheet = workbook.sheet_by_name('Data')
```

```python
N = 29
Ano = zeros(N)
Mag = zeros(N)
Ele = zeros(N)
Ela = zeros(N)
Bar = zeros(N)


for i in range(N):
    Ano[i] = worksheet.cell(0, i).value
    Mag[i] = worksheet.cell(1, i).value
    Ele[i] = worksheet.cell(2, i).value
    Ela[i] = worksheet.cell(3, i).value
    Bar[i] = worksheet.cell(4, i).value
    
```

```python
plt.fill_between(Ano,Bar, label = 'Barocaloric')
plt.fill_between(Ano,Ela+Bar,Bar, label = 'Elastocaloric')
plt.fill_between(Ano,Ela+Bar+Ele,Ela+Bar, label = 'Electrocaloric')
plt.fill_between(Ano,Mag+Ela+Bar+Ele,Ela+Bar+Ele, label = 'Magnetocaloric')
plt.grid()
plt.xlim(1990,2018)
plt.legend(loc = 2)
plt.ylabel('Number of publications')
plt.xlabel('Year')

plt.savefig('Plot_1.pdf',format = 'pdf', dpi=1000,bbox_inches='tight')
```
