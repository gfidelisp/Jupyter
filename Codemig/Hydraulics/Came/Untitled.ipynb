```python
from numpy import *
from scipy import *
import numpy as np
from matplotlib import *

from scipy import integrate
from scipy import interpolate
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import xlrd

rc('text',usetex = True)
rc('font', family='serif',size = 12)
```

```python
Data = xlrd.open_workbook('Testes_valvulateste.xlsx') #Import inputs File
#Inputs = Data.sheet_by_name('Data_new') 
Inputs = Data.sheet_by_name('Inputs')
```

```python
t   = array(Inputs.col_values(0)) #Import AMR Length [mm]
P_up   = array(Inputs.col_values(1)) #Import AMR Height [mm]
P_do   = array(Inputs.col_values(2)) #Import AMR Width [mm]
disp = array(Inputs.col_values(3))#Import temperature span [K]disp_2 = array(Inputs.col_values(4))+273.15 #Import temperature span [K]
disp_2 = array(Inputs.col_values(4))#Import temperature span [K]disp_2 = array(Inputs.col_values(4))+273.15 #Import temperature span [K]
dP = P_up - P_do
```

```python
a = 0
b = 500
```

```python
plt.plot(t[a:b],(P_do[a:b]-np.min(P_do[a:b]))/(np.max(P_do[a:b]) - np.min(P_do[a:b])),'s:', label = 'Pressure')
plt.plot(t[a:b],(P_up[a:b]-np.min(P_up[a:b]))/(np.max(P_up[a:b]) - np.min(P_up[a:b])),'s:', label = 'Pressure')
plt.plot(t[a:b],(disp_2[a:b]-np.min(disp_2[a:b]))/(np.max(disp_2[a:b])-np.min(disp_2[a:b])),'o--', label = 'Position')
#plt.xlim([53.5,55])
plt.grid()
plt.ylabel('Signal [-]')
plt.xlabel('Time [s]')
plt.legend()
plt.savefig('Perfil_hidraulico_2.png',format = 'png', dpi=500,bbox_inches='tight')
```

```python
#plt.plot(t[a:b],P_do[a:b],'o--')
#plt.plot(t[a:b],P_up[a:b],'o--')
plt.plot(t[a:b],dP[a:b],'o--')
plt.plot(t[a:b],disp[a:b]+1,'o--')
```

```python
plt.plot((t[a:b]-t[a])*1000,(dP[a:b]-np.min(dP[a:b]))/(np.max(dP[a:b]) - np.min(dP[a:b])),'s:', label = 'Pressure')
plt.plot((t[a:b]-t[a])*1000,(disp[a:b]-np.min(disp[a:b]))/(np.max(disp[a:b])-np.min(disp[a:b])),'o--', label = 'Position')
#plt.xlim([150,230])
plt.grid()
plt.ylabel('Signal [-]')
plt.xlabel('Time [ms]')
plt.legend()
plt.savefig('Perfil_hidraulico_2.png',format = 'png', dpi=500,bbox_inches='tight')
```
